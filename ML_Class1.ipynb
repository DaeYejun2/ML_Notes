### Multiple 회귀 경사하강법

단일 선형 회귀 | 다중 전형 회귀의 차이

매개변수: $w, b$ | $\vec{w}=[w_1, ..., w_n], b$

모델: $f_{w,b}(x) = wx + b$ | $f_{\vec{w},b}(\vec{x}) = \vec{w} \cdot \vec{x} + b$

비용함수: $J(w, b)$ | $J(\vec{w}, b)$


모든 매개변수 $w_j$와 $b$는 동시에 업데이트 되어야 한다.

$w_j$ 업데이트: $w_j = w_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)})x_j^{(i)}$

$b$ 업데이트: $b = b - \alpha \frac{1}{m} \sum_{i=1}^{m} (f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)})$

## Classification

Binary Classification - 이진 분류

이산적으로 매핑하는 문제

종양이 정상이면 0, 악성이면 1 이런식으로 매핑 - 보통 예측하려는 대상을 1로 지정

임계치를 정해서 분류를 하자. 예를 들어, 임계치가 0.5고, 종양 사이즈가 0.7이면 그냥 악성 이라고 하자 일단.

직선과 임계치가 만나는 부분을 기준으로 어떤 결과로 할지 정할 수 있다.

linear function이 살짝 기울어지게 됨. 악성 파악을 못할 수 있음 -> 데이터 군집에서 멀리 떨어진 데이터 하나 때문에 Decision Boundary가 크게 틀어져 오분류가 발생할 수 있다.

분류의 정답은 0 또는 1인데, 선형 회귀는 0보다 훨씬 작거나 1보다 훨씬 큰 값을 출력할 수 있어 해석이 어렵다.

Logistic Regression = Sigmoid Function

이름은 회귀지만, 실제로는 분류 알고리즘이다. 데이터에 직선이 아닌 **S자 형태의 곡선**을 맞추는 것이 핵심이다.

선형 회귀는 별로 안좋은 듯

S-shaped curve를 그려보자

=> sigmoid function이라고 한다.

선형 모델의 결과값(z)를 0과 1 사이의 값으로 변환해 주는 함수

$$g(z) = \frac{1}{1 + e^{-z}}$$

$z$가 매우 크면 $g(z) \approx 1$, $z$가 매우 작으면 $g(z) \approx 0$, $z=0$이면 $g(z)=0.5$가 된다.

z가 -100 이면 그냥 0으로 가는겨

##### 모델 공식
기존 선형 회귀 식을 시그모이드 함수에 넣는다

$$f_{\vec{w},b}(\vec{x}) = g(\vec{w} \cdot \vec{x} + b) = \frac{1}{1 + e^{-(\vec{w} \cdot \vec{x} + b)}}$$

모델의 출력값 $f_{\vec{w},b}(\vec{x})$는 입력 $x$가 주어졌을 때 $y$가 1일 확률($P(y=1|x)$)로 해석

예를 들어, 출력값이 0.7이면 이 종양이 양성(1)일 확률이 70%다 라는 뜻.

Sigmid 함수의 좋은 점은
0, 1 사이의 값을 낸다.

결국 이런 분류기는 확률을 계산하는 머신, 어디에 매핑될것 이냐.



Decision Boundary: 이 범위를 어떻게 매핑할 것이냐. 또 쓰레시홀드. 0.5 기준으로 참 거짓?  ->

결정 경계는 모델이 0인지 1인지 최종 판단을 내리는 기준선이다.
- 판단 기준: 보통 확률이 0.5 이상이면 1로, 0.5 미만이면 0으로 예측한다.
- 시그모이드 함수 특성상 $z = \vec{w} \cdot \vec{x} + b \ge 0$일 때 확률이 0.5 이상이 된다.
- 따라서 $\vec{w} \cdot \vec{x} + b = 0$이 되는 지점이 두 클래스를 나누는 결정 경계가 된다.

z=0일 때: g(0) = 0.5이기 때문!

정리하자면, 두 클래스를 가르는 기준선은 $\vec{w} \cdot \vec{x} + b = 0$ 인 지점들이 된다.

예를 들어 특성이 두 개($x_1, x_2$)이고 파라미터가 $w_1=1, w_2=1, b=-3$ 이라면:$z = x_1 + x_2 - 3 = 0$

즉, $x_1 + x_2 = 3$ 이라는 직선이 결정 경계가 됩니다.
